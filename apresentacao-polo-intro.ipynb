```python
def hide_code_in_slideshow():
    import os
    uid = os.urandom(8).hex()
    html = """<div id="%s"></div>
    <script type="text/javascript">
        $(function(){
            var p = $("#%s");
            if (p.length==0) return;

            while (!p.hasClass("cell")) {
                p=p.parent();

                if (p.prop("tagName") =="body") return;
            }
            var cell = p;
            cell.find(".input").addClass("hide-in-slideshow")
        });
    </script>""" % (uid, uid)
    display_html(html, raw=True)

def new_section(title):
    style = "text-align:center;background:#2d1970;padding:120px;color:#ffffff;font-size:3em;"
    return HTML('<div style="{}">{}</div>'.format(style, title))

def Images(images, header=None, width="100%",w=['450']*2): # to match Image syntax
    if type(width)==type(1):
        width = "{}px".format(width)
    html = ["""<table style='width:{}'><tr>""".format(width)]
    if header is not None:
        html += ["<th>{}</th>".format(h) for h in header] + ["</tr><tr>"]

    for i,image in enumerate(images):
        html.append("<td><img style='vertical-align:super' src='{}' width='{}'/></td>".format(image,w[i]))
    html.append("</tr></table>")
    display(HTML(''.join(html)))
```
